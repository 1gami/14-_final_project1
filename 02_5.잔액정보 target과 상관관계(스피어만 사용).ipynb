### 잔액정보 target 데이터와의 상관계수

```python
from scipy.stats import pearsonr, pointbiserialr, chi2_contingency
import numpy as np
from scipy.stats import spearmanr
```

```python

# -------------------------------
# Cramér's V 함수
# -------------------------------
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

# -------------------------------
# Point-biserial 함수
# -------------------------------
def point_biserial(x, y):
    return pointbiserialr(x, y)[0]

# -------------------------------
# Segment 기준 상관계수 계산
# -------------------------------
def segment_correlations(df, target_col='Segment'):
    df = df.copy()

    # Segment만 Label Encoding (순서형으로 가정)
    le = LabelEncoder()
    df[target_col] = le.fit_transform(df[target_col])
    target = df[target_col]

    results = {}

    for col in df.columns:
        if col == target_col:
            continue

        x = df[col]

        if pd.api.types.is_numeric_dtype(x):
            # 수치형 → Spearman
            coef, _ = spearmanr(x, target, nan_policy='omit')
            results[col] = coef

        elif x.nunique() == 2:
            # 이진 범주형 → Point-biserial (Label Encoding 필요)
            x_encoded = LabelEncoder().fit_transform(x)
            results[col] = point_biserial(target, x_encoded)

        else:
            # 다중 범주형 → Cramér's V
            results[col] = cramers_v(x, target)

    return pd.Series(results).sort_values(ascending=False)
```

```python
correlations = segment_correlations(df, target_col='Segment')
```

```python
# 상관계수 계산 결과가 correlations에 들어있다고 가정
filtered = correlations.abs()
filtered = filtered[filtered > 0.4].sort_values(ascending=False)
print(filtered)
```
