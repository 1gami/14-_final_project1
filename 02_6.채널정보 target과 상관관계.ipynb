### 6.채널정보 target 데이터와의 상관계수

```python
from scipy.stats import pearsonr, pointbiserialr, chi2_contingency
import numpy as np
```

```python
# 상관계수 계산 함수들 정의
# -------------------------------

# 범주형 변수와의 상관계수 계산 함수
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape)
    return np.sqrt(chi2 / (n * (k - 1))) if k > 1 else np.nan

# 이진분류형 변수와의 상관계수 계산 함수
def point_biserial(x, y):
    return pointbiserialr(x, y)[0]

# 수치형 변수와의 상관계수 계산 함수
def pearson_corr(x, y):
    return pearsonr(x, y)[0]

# -------------------------------
# 상관계수 계산 (Segment 기준)
# -------------------------------

def segment_correlations(df, target_col='Segment'):

    # Step 1. Segment만 Label Encoding
    le = LabelEncoder()
    df[target_col] = le.fit_transform(df[target_col])
    target = df[target_col]

    results = {}

    for col in df.columns:
        if col == target_col:
            continue

        x = df[col]

        if np.issubdtype(x.dtype, np.number):
            # 수치형 변수
            results[col] = pearson_corr(x, target)

        elif x.nunique() == 2:
            # 이진 범주형 → Label Encoding
            x_encoded = LabelEncoder().fit_transform(x)
            results[col] = point_biserial(target, x_encoded)

        else:
            # 다중 범주형 → Cramér’s V
            results[col] = cramers_v(x, target)

    return pd.Series(results).sort_values(ascending=False)
```

```python
correlations = segment_correlations(df, target_col='Segment')
```

```python
# 상관계수 계산 결과가 correlations에 들어있다고 가정
filtered = correlations.abs()
filtered = filtered[filtered > 0.4].sort_values(ascending=False)
print(filtered)
```
