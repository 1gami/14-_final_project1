```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet(r'C:/Users/jk972/OneDrive/Desktop/workspace/final project/git/t마켓팅/8.성과정보_전처리완.parquet')
print(f'전체 shape: {df.shape}')

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print('결측치(수치형) 평균값 대체 완료!')

# 수치형 컬럼 기초 통계
print(df[numeric_cols].describe())

# 주요 변수 분포 시각화
for col in numeric_cols[:5]:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'{col} 분포')
    plt.tight_layout()
    plt.savefig(f'8_{col}_hist.png')
    plt.close()

# EDA 리포트 생성 및 저장 (명세서1 방식)
def make_eda_report(df):
    report = []
    total_count = len(df)
    for col in df.columns:
        dtype = str(df[col].dtype)
        n_null = df[col].isnull().sum()
        n_zero = (df[col] == 0).sum() if pd.api.types.is_numeric_dtype(df[col]) else ''
        n_unique = df[col].nunique()
        desc = df[col].describe() if pd.api.types.is_numeric_dtype(df[col]) else None
        mean = desc['mean'] if desc is not None and 'mean' in desc else ''
        std = desc['std'] if desc is not None and 'std' in desc else ''
        min_ = desc['min'] if desc is not None and 'min' in desc else ''
        max_ = desc['max'] if desc is not None and 'max' in desc else ''
        q25 = desc['25%'] if desc is not None and '25%' in desc else ''
        q50 = desc['50%'] if desc is not None and '50%' in desc else ''
        q75 = desc['75%'] if desc is not None and '75%' in desc else ''
        report.append({
            '변수명': col,
            'dtype': dtype,
            '전체 데이터 개수': total_count,
            'mean': mean,
            'std': std,
            'min': min_,
            '25%': q25,
            '50%': q50,
            '75%': q75,
            'max': max_,
            'n_null': n_null,
            'n_zero': n_zero,
            'n_unique': n_unique
        })
    return pd.DataFrame(report)

eda_report = make_eda_report(df)
eda_report.to_excel('8.성과정보_EDA_report.xlsx', index=False)
print('EDA 리포트 저장 완료!') 
```
